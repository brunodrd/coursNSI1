<h2>2. Implémentation en python de l'algorithme knn </h2>

<h3> Structure </h3>

<ol> Pour impléménter cet algorithme, il nous faut : 
    <li>Une table de données. Cette table peut être une liste ou un dictionnaire. </li>
    <li>Une distance entre deux données.</li>
    <li>Une cible.</li>
</ol>
 

```python


def k_plus_proches_voisins(table, cible, k) :
    """Renvoie la liste des k plus proches voisins de la cible"""
    
    
    def distance_cible(donnee) :
        """ renvoie la distance entre la donnée et la cible"""
        
        return distance(donnee,cibe)
    
    # tri croissant des données suivant la distance à la cible grace au paramètre key
    table_triee = sorted(table, key = distance_cible)
    
    # gestion de la liste des k plus proches voisins
    proches_voisins = []    # initialisation vide
    for i in range(k) : 
        proches_voisins.append(table_triee[i]) # On ajoute les k premières valeurs
        
    return proches_voisins  # On renvoie la liste des $k$ plus proches voisins.
        

```

<h3> L'exercice 2 du cours détaillé</h3>

<h4>L'algorithme knn </h4>
<ul>
	<li><code>Données</code> : </li>
	<ul>
		<li>une <code>table</code> de données de taille $n$ ;</li>
		<li>une donnée cible ;</li>
		<li>un entier $k$ plus petit que $n$ ;</li>
		<li>une règle permettant de calculer la <code>distance</code> entre deux données.</li>
	</ul>
	<li><code>Algorithme</code> : </li>
	<ol>
		<li>Trier les données de la table selon la distance croissante avec la donnée cible.</li>
		<li>Créer la liste des k premières données de la table triée.</li>
		<li>Renvoyer cette liste.</li>
	</ol>
</ul>


```python
# test de la distance de Manhattan

def distance_cible_manhattan(donnee,cible) :
        """ renvoie la distance entre la donnée et la cible, on choisit la distance de Manhattan"""
        
        distance = abs(donnee[1]-cible[0])+abs(donnee[2]-cible[1])
        
        return distance
 

# test de la distance de Tchebychev 

def distance_cible_tchebychev(donnee,cible) :
        """ renvoie la distance entre la donnée et la cible, on choisit la distance de Manhattan"""
        
        distance = max(abs(donnee[1]-cible[0]),abs(donnee[2]-cible[1]))
        
        return distance


donnee=['c1',3,27.2]
cible = [7,28.4]

print("Distance de Manhattan : ", distance_cible_manhattan(donnee,cible))
print("Distance de Tchebychev : ", distance_cible_tchebychev(donnee,cible))
    
```

```python
# On peut utiliser des listes ou des dictionnaires
table = [['t1',1,28], ['t1',3,27.2], ['t1',8,37.6], ['t1',13,40.7], ['t2',2,30], ['t2',3,26], ['t2',10,39], ['t2',15,35.5]]
cible = [7,28.4]
k=3


def k_plus_proches_voisins(table, cible, k) :
    """Renvoie la liste des k plus proches voisins de la cible"""
    
    
    def distance_cible(donnee) :
        """ renvoie la distance entre la donnée et la cible, on choisit la distance de Manhattan"""
        
        distance = abs(donnee[1]-cible[0])+abs(donnee[2]-cible[1])
        
        return distance
    
    table_triee = sorted(table, key = distance_cible)
    
    proches_voisins = []
    
    for i in range(k) : 
        proches_voisins.append(table_triee[i])
        
    return proches_voisins

k_plus_proches_voisins(table,cible,k)
```

```python
import csv
def impor(nom):
    lecteur = csv.DictReader(open(nom + '.csv' , "r", encoding='utf-8-sig' ) , delimiter = ";") # initialisation d’un lecteur de fichier
    return [dict(ligne) for ligne in lecteur]

personnages=impor("personnages")

print(personnages)
```

```python
import csv
```

```python
#fonction qui vient de table
def impor(nom):
    lecteur = csv.DictReader(open(nom + '.csv' , "r", encoding='utf-8') , delimiter = ";") # initialisation d’un lecteur de fichier
    return [dict(ligne) for ligne in lecteur]
impor("choixpeauMagique")
```

```python
poudlard=impor('choixpeauMagique')
```

```python
print(poudlard)
```

```python
def frequence_des_maisons(table) :
    """paramètre: liste de dictionnaires représentatn les élèves
    Renvoie un dictionnaire dont les clés sont les masions et les valeurs les fréquences"""
    frequences={}
    for eleves in table :
        maison=eleves['Maison']
        if maison in frequences.keys():
            frequences[maison]+=1
        else :
            frequences[maison]=1
    return frequences

frequence_des_maisons(poudlard)
```

```python
def maison_majoritaire(table):
    maison=''
    frequences=frequence_des_maisons(table)
    maximum=max(frequences.values())
    for k,val in frequences.items() :
        if val == maximum : maison=k
    return maison
        
maison_majoritaire(poudlard)
```

```python
import csv

#fonction qui vient de table
def impor(nom):
    lecteur = csv.DictReader(open(nom + '.csv' , "r", encoding='utf-8-sig') , delimiter = ";") # initialisation d’un lecteur de fichier
    return [dict(ligne) for ligne in lecteur]
pokemon=impor("pokemon")
```

```python
table=pokemon
k=11
cible={'nom':'Hippodocus','points de vie': 108,'attaque':112,'defense' : 118,'vitesse': 47}


def k_plus_proches_voisins(table, cible, k) :
    """Renvoie la liste des k plus proches voisins de la cible"""
    
    
    def distance_cible(pokemon) :
        """ renvoie la distance entre la donnée et la cible, on choisit la distance de Manhattan"""
        
        distance = abs(eval(pokemon["points de vie"])-cible["points de vie"])+abs(eval(pokemon["attaque"])-cible["attaque"])+abs(eval(pokemon["defense"])-cible["defense"])+abs(eval(pokemon["vitesse"])-cible["vitesse"])
        
        return distance
    
    table_triee = sorted(table, key = distance_cible)
    
    proches_voisins = []
    
    for i in range(k) : 
        proches_voisins.append(table_triee[i])
        
    return proches_voisins

resul=k_plus_proches_voisins(table,cible,k)

def frequence_des_types(table) :
    """paramètre: liste de dictionnaires représentatn les élèves
    Renvoie un dictionnaire dont les clés sont les masions et les valeurs les fréquences"""
    frequences={}
    for pokemon in table :
        types=pokemon['type']
        if types in frequences.keys():
            frequences[types]+=1
        else :
            frequences[types]=1
    return frequences

def type_majoritaire(table):
    type=''
    frequences=frequence_des_types(table)
    maximum=max(frequences.values())
    for k,val in frequences.items() :
        if val == maximum : type=k
    return type

type_majoritaire(resul)
```
